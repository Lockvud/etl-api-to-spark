```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Исходные данные
df_all = pd.read_csv("/home/jovyan/data/users.csv")

# Отфильтрованные (Spark создаёт папку с несколькими .csv)
import os
filtered_path = "/home/jovyan/data/filtered_users.csv"
csv_files = [os.path.join(filtered_path, f) for f in os.listdir(filtered_path) if f.endswith(".csv")]

df_filtered = pd.concat([pd.read_csv(f) for f in csv_files])
```

```python
print(f"Всего пользователей: {len(df_all)}")
print(f"Пользователей с email на .org: {len(df_filtered)}")
```

```python
df_all["domain"] = df_all["email"].apply(lambda x: x.split("@")[1])
df_all["domain"].value_counts().plot(kind="bar", title="Количество пользователей по доменам")
```
